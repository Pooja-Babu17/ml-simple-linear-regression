```python
#pandas is library it contains function related to the database
#it is used for table
#step 1: run pandas help to store all function which undergoes in pandas are storde in our file memorie
#pd is shortform for pandas
import pandas as pd
```

```python
#store the file(datas) in variable(dataset) memory for import
#pd.read_csv is the function
#csv is file format
#in pd class we called the read_csv function
#then give the file name in open peranthesis
#always the file should be in same folder
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# call the variable(dataset)
dataset
```

```python
#split the input
#the input is years of experience it saved dataset
#we have to use two bracket for split the input dat
independent=dataset[["YearsExperience"]]
#call input
independent
```

```python
#split output
dependent=dataset[["Salary"]]
#call output
dependent
```

```python
#split train and test
#sklear is othern library used to split train & test
#model_selection is function name in sklearn
#parameters passed input(independent),output(dependent),test_size ratio train is highet then test
#function are loaded
#write return statement for train & test in x & y , it has 4 list
#x input (train & test) , y output (train & test) 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#train the model
#sklearn provides ml algorithm
#function import  the linear regression
#give variable and assingned the function to that variable
#call fit function
#we have to send train set for learning purose
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#y=wx
#coefficient to calculate and to store
#slope
weight=regressor.coef_
weight
```

```python
#c bais
#intercept variable 
bais=regressor.intercept_
bais
#all saved in regressor
```

```python
#predict is function
#cross check process
y_pred=regressor.predict(X_test)
```

```python
#test model
#we have to pass the actual and predicted output
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#r score is nearly to 1 is good model
#0.9 is nearly to 1 so it good model
r_score
```

```python
#model creation
#use pickle to save model 
#save the finalized model as sav 
import pickle
filename="finslize_model_linear.sav"
```

```python
#model save in regressor, file name,write binary
#it saved as like document
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load the model 
#here we only read there
#then wirte return statement and check the model
loaded_model=pickle.load(open("finslize_model_linear.sav",'rb'))
result=loaded_model.predict([[17]])
```
